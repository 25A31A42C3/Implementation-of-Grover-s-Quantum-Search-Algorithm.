```python
import math
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, transpile, visualization
from qiskit_aer import AerSimulator

def oracle(circuit, solution_string):
    """
    Applies a phase flip to the 'solution_string' state.
    For "101" (qubits 0 and 2 are 1), apply a multi-controlled Z gate.
    Qiskit orders qubits as little-endian (qubit 0 is rightmost).
    """
    if solution_string == "101":
        circuit.cz(0, 2)
    else:
        pass

def inversion_about_mean(circuit, n):
    """
    Applies the Grover Diffusion Operator (inversion about the average).
    Can be implemented as H^n (2|0><0| - I) H^n.
    """
    circuit.h(range(n))
    circuit.z(range(n))
    circuit.cz(0, 1)
    circuit.h(range(n))


n = 3
solution_string = "101"
N = 2**n
M = 1
optimal_num_iterations = math.floor(math.pi / 4 * math.sqrt(N/M) - 0.5)

qr = QuantumRegister(n, "q")
cr = ClassicalRegister(n, "c")
grover_circuit = QuantumCircuit(qr, cr)

grover_circuit.h(range(n))

for i in range(optimal_num_iterations):
    oracle(grover_circuit, solution_string)

grover_circuit.measure(qr, cr)

simulator = AerSimulator()
transpiled_circuit = transpile(grover_circuit, simulator)
result = simulator.run(transpiled_circuit, shots=1024).result()
counts = result.get_counts(grover_circuit)

print(f"Measurement results: {counts}")

```
